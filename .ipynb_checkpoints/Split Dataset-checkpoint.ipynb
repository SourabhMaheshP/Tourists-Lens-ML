```python
from DLModel.CNN_Model import Cnn
import os
import shutil
import random
```

```python
TRAIN_PATH = r'C:\Users\mahesh\TouristsLens\Model\Dataset\train'
TEST_PATH = r'C:\Users\mahesh\TouristsLens\Model\Dataset\test'

size = Cnn.datasetSize(TRAIN_PATH)
category_size = len(Cnn.categoryList(r'C:\Users\mahesh\TouristsLens\Model\Categories.txt'))
print(size, category_size)
```

```python
# Dataset split Ratio 80 : 20
train_size = (size * 8)//10 
test_size = size - train_size
print(train_size , test_size)
```

```python
def moveFile(src, dest):
    if not os.path.exists(dest):
        os.mkdir(dest) 
    shutil.move(src,dest)
    

class_size = test_size//category_size
# print(class_size)
classes = os.listdir(TRAIN_PATH)
# print(classes)
for name in classes : 
    non_repeat = []
    file_path = os.path.sep.join([TRAIN_PATH,name])
    _,_,files = next(os.walk(file_path))
    files = files[:-1]
#     print(len(files))
    L = len(files)
    for i in range(1,class_size+1):
        index = random.randint(1,L)
        while index in non_repeat :
            index = random.randint(1,L)
        non_repeat.append(index)
        print(index)
        # non-repeating random index 
        moveFile(os.path.sep.join([file_path,files[index]]),os.path.sep.join([TEST_PATH,name])) 
        L -= 1
    non_repeat.clear()
```
